```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

## Figure 3.3
Gibbs energy G versus reaction extent $\epsilon$ 

```python
# equilibrium const
K=108
# pressure
P=2.5
npts=100
# grid of reaction extents
x=np.linspace(0,0.5,npts)
# reaction extents except 0 and 0.5 to prevent log evaluation errors
xsub=x[1:(len(x)-1)]
# Gibbs free energy initialization
G=np.zeros(npts)
# Gibbs free energy at reaction extent >0 and < 0.5
G[1:(len(x)-1)] = -xsub*log(108) + (1-xsub)*log(2.5) + xsub*log(xsub) + 2*(0.5-xsub)*log(0.5-xsub) - (1-xsub)*log(1-xsub)
# Gibbs free energy at reaction extent 0 and 0.5
G[0]=log(2.5) + 2*0.5*log(0.5)
G[npts-1]=-0.5*log(108) + 0.5*log(2.5)
```

```python
fig=plt.figure()

axes=fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x,G,'b')
axes.set_xlabel(r'$\epsilon$',fontsize=20)
axes.set_ylabel(r'$\tilde{G}$',fontsize=20)

axes2=fig.add_axes([0.5,0.5,0.4,0.4])
axes2.plot(x,G,'b')
axes2.set_xlim([0.45,0.5])
axes2.set_ylim([-1.95,-1.88])
    
```
